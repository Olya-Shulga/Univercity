#### **1. Обоснование необходимости разработки**


Для кого предназначено разрабатываемое Вами решение и какие потребности целевой аудитории оно призвано удовлетворить?




*Обоснуйте актуальность решаемой проблемы для конкретных заинтересованных сторон.*

<font color='red'><b>Ваш ответ</b></font>


Разрабатываемое решение предназначено для специалистов в области дистанционного зондирования Земли (ДЗЗ), включая:
* Научные организации (институты РАН, университетские лаборатории), занимающиеся мониторингом земного покрова
* Коммерческие компании в сфере агротехнологий, лесного хозяйства, градостроительства
* Государственные ведомства (МЧС, Росреестр, Минприроды) для оперативного анализа территорий

Потребности целевой аудитории:
* Обработка крупноформатных спутниковых снимков (4096×4096 и выше) без разбиения на тайлы
* Снижение вычислительных затрат при сохранении точности сегментации
* Возможность развёртывания на оборудовании с ограниченными ресурсами (полевые условия, БПЛА)

Актуальность проблемы:
Современные трансформер-архитектуры (ViT, Swin) имеют квадратичную сложность O(N²) по памяти и вычислениям, что делает обработку снимков высокого разрешения экономически и технически затратной. Архитектура Mamba обеспечивает линейную сложность O(N), а межслойное накопление признаков (по аналогии с HAViT) улучшает точность за счёт сохранения контекста между слоями. На момент начала работы (2026 г.) данная комбинация не была применена к задачам сегментации ДЗЗ, что подтверждает научную новизну.
