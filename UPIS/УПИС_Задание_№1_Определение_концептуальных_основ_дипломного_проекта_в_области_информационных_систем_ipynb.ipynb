#### **3. Методология использования**


Каким образом предполагается практическое применение разрабатываемого решения? Опишите основные сценарии взаимодействия пользователей с разрабатываемым решением.

Сценарий практического применения:

Разрабатываемое решение внедряется как модуль в существующие ГИС-системы или используется как автономное ПО для пакетной обработки снимков.

Основные сценарии взаимодействия:
1. Загрузка данных: Пользователь загружает пакеты спутниковых снимков и, при наличии, размеченные данные для валидации.
2. Выбор конфигурации: Через CLI или GUI пользователь выбирает архитектуру (например, Channel2DTransformer для мультиспектральных данных) и стратегию обучения (Full-training или Prompt-based tuning).
3. Запуск обработки: Система выполняет предобработку (нормализация, аугментация), запускает нейросеть и формирует карты классов или bounding-boxes.
4. Экспорт результатов: Пользователь получает векторные слои или растровые маски, совместимые с QGIS/ArcGIS.

________________________________________________________________________________
Что происходит внутри системы (Выбор конфигурации):

Парсинг конфигурации: Система считывает параметры и инициализирует соответствующие модули.

Выбор архитектуры:
* RSS — если задача связана с реконструкцией или сверхразрешением.
* Channel2DTransformer — если важны межканальные зависимости (мультиспектральный анализ).
* PromptMatcher — если мало размеченных данных и нужна быстрая адаптация.

Настройка стратегии обучения:
* Full-training: все веса модели обучаются с нуля или fine-tuning.
* Prompt-based tuning: "замораживается" основная модель, обучаются только prompt-векторы (значительно меньше параметров).
* Few-shot: модель адаптируется на 5–20 примерах нового класса.
