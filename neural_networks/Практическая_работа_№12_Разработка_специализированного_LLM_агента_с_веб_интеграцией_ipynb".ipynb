```python
import os
import time
import shutil
from datetime import datetime

import gradio as gr

from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_chroma import Chroma
from langchain_ollama import OllamaEmbeddings, ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

from tavily import TavilyClient
```

```python
MODEL_NAME = "qwen3.5:9b"
EMBEDDING_MODEL = "nomic-embed-text-v2-moe:latest"

CHUNK_SIZE = 800
CHUNK_OVERLAP = 150
CHROMA_PATH = ".chromadb_excuse_agent"
CHAT_HISTORY = []
PDF_RETRIEVER = None

AGENT_NAME = "ExcuseGPT — генератор уважительных оправданий"

AGENT_ROLE = """
Ты — ExcuseGPT, ироничный ИИ-агент, который помогает придумывать правдоподобные,
вежливые и слегка смешные оправдания для разных жизненных ситуаций.

Твоя специализация:
- оправдания для дедлайнов;
- объяснения опозданий;
- вежливые ответы преподавателю;
- мягкие формулировки для работы;
- юмористические, но не грубые объяснения.

Правила:
1. Не предлагай опасные, незаконные или вредные действия.
2. Не советуй откровенно обманывать, если можно сформулировать мягко и честно.
3. Давай 2–3 варианта ответа: официальный, нейтральный и смешной.
4. Если вопрос связан с учебой или работой, сохраняй уважительный тон.
5. Отвечай на русском языке.
"""
```

```python
def load_pdf(pdf_path):
    loader = PyMuPDFLoader(pdf_path)
    documents = loader.load()
    print(f"Загружено страниц: {len(documents)}")
    return documents

def split_into_chunks(documents, chunk_size=CHUNK_SIZE, chunk_overlap=CHUNK_OVERLAP):
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
        length_function=len,
        separators=["\n\n", "\n", ".", " ", ""]
    )
    chunks = splitter.split_documents(documents)
    print(f"Создано чанков: {len(chunks)}")
    return chunks

def create_vectorstore(chunks):
    embeddings = OllamaEmbeddings(model=EMBEDDING_MODEL)

    vectorstore = Chroma.from_documents(
        documents=chunks,
        embedding=embeddings
    )

    print("Векторная база создана в памяти")
    return vectorstore

def process_pdf(pdf_path):
    documents = load_pdf(pdf_path)
    chunks = split_into_chunks(documents)
    vectorstore = create_vectorstore(chunks)
    retriever = vectorstore.as_retriever(search_kwargs={"k": 4})
    return vectorstore, retriever

def format_documents(docs):
    if not docs:
        return "Контекст из PDF отсутствует."
    return "\n\n".join([doc.page_content for doc in docs])
```

```python
def search_web_tavily(query, max_results=3):
    try:
        result = tavily_client.search(query=query, max_results=max_results)
        items = result.get("results", [])
        if not items:
            return "Интернет-поиск ничего не нашёл."

        parts = []
        for i, item in enumerate(items, 1):
            title = item.get("title", "Без названия")
            content = item.get("content", "")
            url = item.get("url", "")
            parts.append(f"{i}. {title}\n{content}\nИсточник: {url}")
        return "\n\n".join(parts)
    except Exception as e:
        return f"Ошибка Tavily: {e}"
```

```python
def history_to_text():
    if not CHAT_HISTORY:
        return "История диалога пока пуста."
    blocks = []
    for i, (user_msg, bot_msg) in enumerate(CHAT_HISTORY, 1):
        blocks.append(f"Диалог {i}\nПользователь: {user_msg}\nАгент: {bot_msg}")
    return "\n\n".join(blocks)

def save_history_to_txt(filename="chat_history.txt"):
    text = history_to_text()
    with open(filename, "w", encoding="utf-8") as f:
        f.write(f"{AGENT_NAME}\n")
        f.write(f"Дата: {datetime.now()}\n\n")
        f.write(text)
    return filename
```

```python
def create_rag_chain():
    llm = ChatOllama(
        model=MODEL_NAME,
        temperature=0.3
    )

    prompt = ChatPromptTemplate.from_template("""
Ты должен ответить как специализированный агент.

Роль агента:
{agent_role}

История диалога:
{history}

Контекст из PDF:
{pdf_context}

Контекст из интернета:
{web_context}

Вопрос пользователя:
{question}

Правила ответа:
- Отвечай по-русски.
- Отвечай понятно и дружелюбно.
- Если пользователь просит оправдание, предложи 3 варианта:
  1. Официальный.
  2. Нейтральный.
  3. Смешной.
- Если пользователь просит написать сообщение преподавателю или начальнику, предложи готовый текст сообщения.
- Если есть полезный контекст из PDF, используй его.
- Если есть актуальная информация из интернета, используй её.
- Не выдумывай факты, если данных нет.
- Сохраняй уважительный тон даже в юмористических ответах.
""")

    chain = prompt | llm | StrOutputParser()
    return chain

rag_chain = create_rag_chain()
```

```python
def should_use_web(question: str) -> bool:
    markers = [
        "сейчас", "сегодня", "новости", "актуально", "в интернете",
        "найди", "поиск", "статьи", "советы", "эксперты",
        "этикет", "деловая переписка", "письмо", "как написать",
        "преподавателю", "начальнику", "работодателю",
        "университет", "работа", "дедлайн", "опоздание"
    ]
    q = question.lower()
    return any(m in q for m in markers)


def generate_answer(question):
    global PDF_RETRIEVER

    history_text = history_to_text()

    pdf_context = "PDF не загружен."
    if PDF_RETRIEVER is not None:
        try:
            docs = PDF_RETRIEVER.invoke(question)
            pdf_context = format_documents(docs)
        except Exception as e:
            pdf_context = f"Ошибка получения контекста из PDF: {e}"

    web_context = "Интернет-поиск не использовался."
    if should_use_web(question):
        web_context = search_web_tavily(question, max_results=3)

    answer = rag_chain.invoke({
        "agent_role": AGENT_ROLE,
        "history": history_text,
        "pdf_context": pdf_context,
        "web_context": web_context,
        "question": question
    })

    CHAT_HISTORY.append((question, answer))
    return answer
```

```python
def chat_fn(message, history):
    answer = generate_answer(message)
    history = history + [
        {"role": "user", "content": message},
        {"role": "assistant", "content": answer}
    ]
    return history, ""


def save_fn():
    filename = save_history_to_txt("excusegpt_history.txt")
    return f"История сохранена в файл: {filename}"


with gr.Blocks() as demo:
    gr.Markdown("# ExcuseGPT — генератор уважительных оправданий")
    gr.Markdown(
        """
        Этот ИИ-агент помогает придумывать вежливые, нейтральные и немного смешные оправдания
        для учебы, работы и повседневных ситуаций.

        Можно попросить:
        - написать сообщение преподавателю;
        - объяснить опоздание;
        - придумать оправдание для дедлайна;
        - сформулировать ответ начальнику;
        - сделать официальный, нейтральный или смешной вариант.
        """
    )

    chatbot = gr.Chatbot(type="messages")

    msg = gr.Textbox(
        label="Твой запрос",
        placeholder="Например: помоги написать преподавателю, что я не успела сдать практическую"
    )

    send_btn = gr.Button("Отправить")
    save_btn = gr.Button("Сохранить историю")
    save_output = gr.Textbox(label="Результат сохранения")

    gr.Examples(
        examples=[
            "Помоги написать преподавателю сообщение о том, что я не успела сдать практическую работу вовремя",
            "Придумай смешное, но приличное оправдание, почему я опоздала на пару",
            "Как объяснить начальнику задержку отчета?",
            "Составь официальное оправдание, почему я не успела выполнить задачу к дедлайну",
            "Придумай нейтральное и уважительное сообщение о переносе срока"
        ],
        inputs=msg
    )

    send_btn.click(chat_fn, inputs=[msg, chatbot], outputs=[chatbot, msg])
    msg.submit(chat_fn, inputs=[msg, chatbot], outputs=[chatbot, msg])
    save_btn.click(save_fn, inputs=[], outputs=save_output)

demo.launch(debug=True, share=True)
```
